```python
from aimoppo import AIMOPPOTrainer


solver = AIMOPPOTrainer(
    "configs/costar_cot_1shot.yaml",
    inference=True,
)
```

```python
result = solver.infer("Let $k, l > 0$ be parameters. The parabola $y = kx^2 - 2kx + l$ intersects the line $y = 4$ at two points $A$ and $B$. These points are distance 6 apart. What is the sum of the squares of the distances from $A$ and $B$ to the origin?")
result.show_text()
```
